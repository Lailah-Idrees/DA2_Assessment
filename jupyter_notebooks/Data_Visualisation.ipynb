Python had issues running the swarmplot with my data; The issue: When there are too many data points or the markers are too large, seaborn cannot find non-overlapping positions for all points. The percentages indicate how many points couldn't be placed:

16.9% in the first plot
13.0% in the second plot
33.4% in the third plot (most severe)
21.0% in the fourth plot

so coppilot suggested using a striplot which worked out much better
