# Section 2

Distribution of Potability 

```python
# Convert 0/1 into readable labels BEFORE plotting
df["Potability"] = df["Potability"].map({
    0: "Not potable",
    1: "Potable"
})

# Now calculate proportions
Potability_count = df.value_counts('Potability', normalize=True)

# Plot the pie chart
plt.pie(
    x=Potability_count,
    labels=Potability_count.index,
    autopct='%1.1f%%'
)
plt.show()
```

Bivariate Analysis: Comparing the Impact of different features on water potability 

```python
x,y= 'ph','Potability'
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

sns.boxplot(data=df, x=x, y=y, ax=axes[0])
sns.swarmplot(data=df, x=x, y=y, ax=axes[1], dodge=True)
plt.show()
```

```python
sns.lmplot(data=df, x='Hardness', y='ph', hue='Potability', height=5, aspect=1.5)

plt.show()
```

```python
sns.set_theme(style="whitegrid")
sns.barplot(x='Potability', y='ph', data=df)
plt.title('pH by Potability')
```
